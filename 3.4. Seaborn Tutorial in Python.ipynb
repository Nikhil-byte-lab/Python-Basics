Seaborn:

    Data Visualization Library

Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
house_cali = fetch_california_housing()

house = pd.DataFrame(house_cali.data,columns=house_cali.feature_names)
house['PRICE'] = house_cali.target
```

```python
print(house_cali)
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation:
    
    1. Postive Correlation
    2. Negative Correlation

Heat Map

```python
correlation = house.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
